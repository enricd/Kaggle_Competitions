```python
# TTA Strtegy ordered queues merge (worse strategy)

model1.cuda()
dl1 = dm1.test_dataloader()
labels_epochs = []
observations = []
for i in tqdm(range(5)):
    labels = []
    for batch in tqdm(dl1):
        preds = model1.predict(batch)
        values, ixs = preds.topk(30)
        labels += ixs.tolist()
        if i==0:
            observation_ids = batch["observation_id"]
            observations += observation_ids.numpy().tolist()
    labels_epochs.append(labels)
```

```python
def top30unique(l):
    ltop = []
    i = 0
    while len(ltop) < 30:
        if l[i] not in ltop:
            ltop.append(l[i])
        i += 1
        
    return ltop
```

```python
labels_merged = []
for i in tqdm(range(len(observations))):
    obs = []
    for j in range(30):
        for e in labels_epochs[0:5]:
            obs.append(e[i][j])
    top_obs = top30unique(obs)
    labels_merged.append(" ".join([str(i) for i in top_obs]))
```

```python
model1.cuda()
all_probas = []
batch_probas = []
observations = []
dl1 = dm1.test_dataloader()
for i in tqdm(range(5)):
    probas = torch.tensor([])
    for batch in tqdm(dl1):
        preds = model1.predict(batch)
        probas = torch.concat([probas, preds.cpu()], dim=0)
        if i == 0:
            observation_ids = batch["observation_id"]
            observations += observation_ids.cpu().numpy().tolist()
    batch_probas.append(probas)
```

```python
_batch_probas = torch.stack(batch_probas, dim=0)
_batch_probas = _batch_probas.mean(dim=0)
all_probas.append(_batch_probas)
```
