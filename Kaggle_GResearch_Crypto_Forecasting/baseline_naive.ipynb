```python
import os
from pathlib import Path

path = Path("data")
os.listdir(path)
```

```python
import pandas as pd

train = pd.read_csv(path / "train.csv")
train = train.sort_values(by=["timestamp", "Asset_ID"])
train.index = pd.to_datetime(train["timestamp"], unit="s")

train
```

```python
train[train["Asset_ID"] == 1].index[-10:].values
```

```python
test = pd.read_csv(path / "example_test.csv")
test = test.sort_values(by=["timestamp", "Asset_ID"])
test.index = pd.to_datetime(test["timestamp"], unit="s")

test[test["Asset_ID"] == 1]
```

```python
sup = pd.read_csv(path / "supplemental_train.csv")
sup = sup.sort_values(by=["timestamp", "Asset_ID"])
sup.index = pd.to_datetime(sup["timestamp"], unit="s")

sup[sup["Asset_ID"] == 1]
```

```python
sup[sup["Asset_ID"] == 1][-100:].Target.plot()
```

```python
kk = sup[sup["Asset_ID"] == 1][-100:]
kk.Target2 = kk.Target.fillna(0)
kk.Target2.plot()
```

```python
kk.Target3 = kk.groupby("Asset_ID").Target.fillna(method="ffill")
kk.Target3.plot()
```

```python
clean_target = sup.groupby("Asset_ID").ffill()
clean_target["Asset_ID"] = sup["Asset_ID"].values.copy()
last_values = clean_target[["Asset_ID", "Target"]][-14:]

last_values
```

```python
submission = pd.read_csv(path / "example_sample_submission.csv")
submission
```

```python
for _, row in submission.iterrows():
    r = row["row_id"]
    asset_id = test.loc[test["row_id"] == r]["Asset_ID"].item()
    target = last_values.loc[last_values["Asset_ID"] == asset_id]["Target"].item()
    submission.loc[submission["row_id"] == r, "Target"] = target
```

```python
# Kaggle notebook version of last cell

import gresearch_crypto

env = gresearch_crypto.make_env()
iter_test = env.iter_test()

for i, (test_df, sample_prediction_df) in enumerate(iter_test):
    for _, row in test_df.iterrows():
        try:
            target = last_values.loc[last_values["Asset_ID"] == row["Asset_ID"], "Target"].item()
            sample_prediction_df.loc[sample_prediction_df["row_id"] == row["row_id"], "Target"] = target
        except:
            sample_prediction_df.loc[sample_prediction_df["row_id"] == row["row_id"], "Target"] = 0. 
    env.predict(sample_prediction_df)
```
